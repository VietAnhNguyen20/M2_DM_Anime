```python
username = []
community = []
for n in G.nodes:
  username.append(n)
  community.append(G.nodes.get(n)["Modularity Class"])


```

```python
import pandas as pd

df = pd.DataFrame.from_dict({"username":username, "community":community})
print(len(df))
df.head(5)
```

```python
users_df = pd.read_csv("./Data/users_cleaned.csv")
users_df = users_df.reset_index(drop=True)
users_df.head(5)
```

```python
df = df.merge(users_df, on="username")
df
```

```python
# Fonction de viet anh
from datetime import datetime
def get_time_period(begin_time):
    #2018 is the year that this dataset is collected
    dob_date = datetime.strptime(begin_time, "%Y-%m-%d %H:%M:%S")
    current_date = datetime.now()
    years = 2018 - dob_date.year - ((current_date.month, current_date.day) < (dob_date.month, dob_date.day))
    return years
```

```python
import matplotlib.colors as mcolors

age_com0 = [get_time_period(a) for a in community0["birth_date"].to_list()]
age_com1 = [get_time_period(a) for a in community1["birth_date"].to_list()]

pyplot.hist(age_com1, alpha=0.5, label='Community 1', color="crimson")
pyplot.hist(age_com0, alpha=0.5, label='Community 0', color="green")
pyplot.legend(loc='upper right')
pyplot.show()
```
