```python
# Fonction de viet anh
from datetime import datetime
def get_time_period(begin_time):
    #2018 is the year that this dataset is collected
    dob_date = datetime.strptime(begin_time, "%Y-%m-%d %H:%M:%S")
    current_date = datetime.now()
    years = 2018 - dob_date.year - ((current_date.month, current_date.day) < (dob_date.month, dob_date.day))
    return years
```

```python
import matplotlib.colors as mcolors

age_com0 = [get_time_period(a) for a in community0["birth_date"].to_list()]
age_com1 = [get_time_period(a) for a in community1["birth_date"].to_list()]

pyplot.hist(age_com1, alpha=0.5, label='Community 1', color="crimson")
pyplot.hist(age_com0, alpha=0.5, label='Community 0', color="green")
pyplot.legend(loc='upper right')
pyplot.show()
```

```python
listAnime = list(set(animewatched_com0).union(set(animewatched_com1)))
countCom0 = {}
countCom1 = {}
for a in listAnime:
  countCom0[a] = 0
  countCom1[a] = 0

print(len(animewatched_com0))
print(len(animewatched_com0))


for a in animewatched_com0:
  countCom0[a]+=1
for a in animewatched_com1:
  countCom1[a]+=1

for i in countCom0:
  countCom0[i] = (countCom0[i] * 100 / len(animewatched_com0))
for i in countCom1:
  countCom1[i] = (countCom1[i] * 100 / len(animewatched_com1))
```

```python
import numpy as np
diff = [x for x in np.subtract(list(countCom0.values()), list(countCom1.values()))]
```
