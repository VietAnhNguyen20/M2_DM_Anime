```python
anime = pd.read_csv('./Data/anime_cleaned.csv')
username_and_anime_id = username_and_anime_id.merge(anime, on="anime_id")
```

```python
username_and_anime_id = username_and_anime_id[["username", "genre"]]
username_and_anime_id
```

```python
tmp = [[y.replace(" ", "") for y in x.split(",")] for x in username_and_anime_id["genre"]]
username_and_anime_id["genre"] = tmp
username_and_anime_id
```

```python
genreWatched = username_and_anime_id.explode("genre")
genreWatched
```

```python
genreList = genreWatched["genre"].unique().tolist()
print(len(genreList))
genreList
```

```python
genreList_T = genreWatched.pivot(columns='genre', values="username")
genreList_T=genreList_T.notnull().astype("int")
genreList_T
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

supports = apriori(genreList_T, min_support=0.1,max_len=3, use_colnames=True)
supports.sort_values(by="support", ascending=False).head(50)
```

```python
# compute association rules
rules = association_rules(supports, min_threshold=0.5)
rules.sort_values(by="conviction", ascending=True).head(30)
```
