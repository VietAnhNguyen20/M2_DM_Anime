```python
import pandas as pd
from collections import defaultdict
from tqdm import tqdm
import random

random.seed(0)
```

```python
# nb nodes, the n biggest anime watcher
NB_USERS = 500
```

```python
medianWatcherDf = pd.read_csv("./Data/animeList_selected.csv", index_col=0)
medianWatcherDf = medianWatcherDf.reset_index(drop=True)
medianWatcherDf['nbAnimeWatched'] = medianWatcherDf.groupby('username')['username'].transform('count')

medianWatcherDf = medianWatcherDf.sort_values(by='nbAnimeWatched', ascending=False)

print(medianWatcherDf['nbAnimeWatched'].median())

biggestWatcher = medianWatcherDf['username'].unique()
biggestWatcher = biggestWatcher.tolist()
```

```python
print(biggestWatcher.index('_Ami_'))
print(biggestWatcher[29112-100:29112+100])


biggestWatcherCenteredOnMedian = biggestWatcher[29112-20000:29112+20000:80]

biggestWatcherCenteredOnMedianSample = biggestWatcherCenteredOnMedian
```

```python
anime_selected = pd.read_csv("./Data/animeList_selected.csv", index_col=0)
anime_selected = anime_selected.reset_index(drop=True)

# Filter
anime_selected = anime_selected[anime_selected['username'].isin(biggestWatcherCenteredOnMedianSample)]


anime_selected.head(5)
```

```python
username_and_anime_id = anime_selected.drop(columns=["my_watched_episodes", "my_score", "my_status", "my_rewatching", "my_rewatching_ep", "my_tags"])
```

```python
username_and_anime_id
```

```python
list_user_id = anime_selected['username'].unique().tolist()
print(f"Number of user_id: {len(list_user_id)}")

#list_user_id = random.sample(list_user_id,len(list_user_id)//100)

print(f"Number of user_id: {len(list_user_id)}")
```

```python
# take 5-6 min

animeToUserList = defaultdict(list)

for idx, row in tqdm(username_and_anime_id.iterrows()):
    if row["username"] in list_user_id:
        animeToUserList[row["anime_id"]].append(row["username"])
    

animeToUserList
```

```python
import numpy as np
adj = np.zeros((len(list_user_id),len(list_user_id)))
count = 0
for userlist in tqdm(animeToUserList.values()):
  for i in range(len(userlist)):
    for j in range(i+1, len(userlist)):
      try:
        adj[list_user_id.index(userlist[i]), list_user_id.index(userlist[j])] += 1
      except:
        pass

```

```python
import networkx as nx
G = nx.Graph()



for id1 in range(len(adj)):
  for id2 in range(len(adj)):
    #print(id1,id2,end="\r")
    
    list_user_id[id1]
    
    if adj[id1][id2] >= 50:
      G.add_node(list_user_id[id1])
      G.add_node(list_user_id[id2])
      G.add_edge(list_user_id[id1], list_user_id[id2], weight=adj[id1][id2])

nx.write_graphml(G, "userGraph.graphml")
```
