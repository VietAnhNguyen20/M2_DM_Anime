```python
anime_selected = pd.read_csv("./Data/animeList_selected.csv", index_col=0)
anime_selected = anime_selected.reset_index(drop=True)
anime_selected.head(5)
```

```python
username_and_anime_id = anime_selected.drop(columns=["my_watched_episodes", "my_score", "my_status", "my_rewatching", "my_rewatching_ep", "my_tags"])
```

```python
list_user_id = anime_selected['username'].unique().tolist()
print(f"Number of user_id: {len(list_user_id)}")

list_user_id = random.sample(list_user_id,len(list_user_id)//100)

print(f"Number of user_id: {len(list_user_id)}")
```

```python
# take 5-6 min

animeToUserList = defaultdict(list)

for idx, row in tqdm(username_and_anime_id.iterrows()):
    if row["username"] in list_user_id:
        animeToUserList[row["anime_id"]].append(row["username"])
    

animeToUserList
```

```python
import numpy as np
adj = np.zeros((len(list_user_id),len(list_user_id)))
count = 0
for userlist in tqdm(animeToUserList.values()):
  for i in range(len(userlist)):
    for j in range(i+1, len(userlist)):
      try:
        adj[list_user_id.index(userlist[i]), list_user_id.index(userlist[j])] += 1
      except:
        pass

```
